```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
```

```python
c1 = '#c0791b'
c2 = '#1b786e'
c3 = '#599191'
c4 = '#5f5f5f'
c5 = '#202112'
```

```python
df = pd.read_csv("../data/chatgpt_short_clean_combined_en.csv", 
                 parse_dates=["at", "repliedAt"])
```

```python
for var in ['Source', 'score', 'appVersion']:
    list=[]
    for i in range(1,6,1):
        x = df[var].unique()[i-1] if df[var].nunique() >= i else ''
        list.append(x)
    fig = px.histogram(df.sort_values(var), 
        x=var,
        title="Number of Reviews by " + var.capitalize(),
        labels={'count': 'Reviews'},
        text_auto=True,
        template='ggplot2',
        hover_data={"at_ym": "|%B %d, %Y"},
        color=var,
        color_discrete_map={
            list[0]: c1, 
            list[1]: c2, 
            list[2]: c3, 
            list[3]: c4,     
            list[4]: c5
            })
    fig.update_traces(textposition='auto')
    fig.update_layout(legend_title_text=var.capitalize())
    fig.update_xaxes(
        dtick="M1",
        tickformat="%B\n%Y")
    fig.update_layout(
       yaxis_title="Reviews")
    fig.show()
```

```python
fig,ax = plt.subplots(2,3,figsize=(16,16))
for var in ['Source', 'score', 'appVersion']:
    df_chart = df.groupby([var]).size().reset_index(name='Reviews') 
    df_chart['Reviews (%)'] = df_chart['Reviews'].transform(lambda x: x / x.sum() * 100)
    list=[]
    for i in range(1,6,1):
        x = df[var].unique()[i-1] if df[var].nunique() >= i else ''
        list.append(x)
    fig = px.histogram(df_chart.sort_values(var), 
        y='Reviews (%)', 
        x=var,
        title="Proportion of Reviews by " + var.capitalize(),
        labels={'count': 'Reviews'},
        text_auto="2.1f",
        template='ggplot2',
        color=var,
        color_discrete_map={
            list[0]: c1, 
            list[1]: c2, 
            list[2]: c3, 
            list[3]: c4,     
            list[4]: c5
            })
    fig.update_traces(
        textposition='auto')
    fig.update_layout(legend_title_text=var.capitalize())
    fig.update_layout(
       yaxis_title="Reviews")
    fig.show()
```

```python
# Filtered for Google

for var in ['score', 'appVersion', 'reply']:
    df_chart = df[df.Source == 'Google'].groupby(['at',var]).size().reset_index(name='Reviews')
    list=[]
    for i in range(1,6,1):
        x = df[var].unique()[i-1] if df[var].nunique() >= i else ''
        list.append(x)
    fig = px.line(df_chart, 
        y='Reviews', 
        x='at',
        title="Daily Number of Reviews by " + var.capitalize(),
        labels={'at': 'Review Date'},
        template='ggplot2',
        hover_data={"at": "|%B %d, %Y"},
        color=var,
        color_discrete_map={
            list[0]: c1, 
            list[1]: c2, 
            list[2]: c3, 
            list[3]: c4,     
            list[4]: c5
            })
    fig.update_layout(legend_title_text=var.capitalize())
    fig.update_xaxes(
        dtick="M1",
        tickformat="%d.%B\n%Y")
    fig.show()
```

```python
for var in ['score', 'reply']:
    df_chart = df[df.Source == 'Google'].groupby(['at',var]).size().reset_index(name='Reviews') 
    df_chart['Reviews (%)'] = df_chart.groupby('at')['Reviews'].transform(lambda x: x / x.sum() * 100)
    list=[]
    for i in range(1,6,1):
        x = df[var].unique()[i-1] if df[var].nunique() >= i else ''
        list.append(x)
    fig = px.line(df_chart, 
        y='Reviews (%)', 
        x='at',
        title="Proportion of Daily Reviews by " + var.capitalize(),
        labels={'at': 'Review Date'},
        template='ggplot2',
        hover_data={"at": "|%B %d, %Y"},
        color=var,
        color_discrete_map={
            list[0]: c1, 
            list[1]: c2, 
            list[2]: c3, 
            list[3]: c4,     
            list[4]: c5
            })
    fig.update_layout(legend_title_text=var.capitalize())
    fig.update_xaxes(
        dtick="M1",
        tickformat="%d.%B\n%Y")
    fig.show()
```

```python
for var in ['score']:
    for s in ['Apple', 'Google']:
        df_chart = df[df.Source == s].groupby(['at_ym',var]).size().reset_index(name='Reviews') 
        df_chart['Reviews (%)'] = df_chart.groupby('at_ym')['Reviews'].transform(lambda x: x / x.sum() * 100)
        max_cat = df_chart[var].iloc[np.argmax(df_chart['Reviews (%)'])]
        print(max_cat)
        #df_chart['labels'] = np.where(df_chart[var]==max_cat,df_chart['Reviews (%)'], "")
        list=[]
        for i in range(1,6,1):
            x = df[var].unique()[i-1] if df[var].nunique() >= i else ''
            list.append(x)
        fig = px.line(df_chart, 
            y='Reviews (%)', 
            x='at_ym',
            text = np.where(df_chart[var]==max_cat,df_chart['Reviews (%)'], " "),
            title="Proportion of Monthly " + s + " Reviews by " + var.capitalize(),
            labels={'at_ym': 'Review Month'},
            template='ggplot2',
            hover_data={"at_ym": "|%B %d, %Y"},
            color=var,
            color_discrete_map={
                list[0]: c1, 
                list[1]: c2, 
                list[2]: c3, 
                list[3]: c4,     
                list[4]: c5
                })
        fig.update_traces(
            textposition='bottom center',
            texttemplate= "%{text:2.1f}%",
            selector = ({'name':str(max_cat)}))
        fig.update_layout(
            legend_title_text=var.capitalize(),
            autosize=False,
            width=700,
            height=400,
            font=dict(
                #family="Courier New, monospace",
                size=14,  # Set the font size here
                #color="RebeccaPurple"
            )
            )
        fig.update_xaxes(
            dtick="M1",
            tickformat="%B\n%Y")
        fig.show()
```

```python
for var in ['Source', 'appVersion', 'reply']:
    df_chart = df.groupby(['at_ym',var]).size().reset_index(name='Reviews')
    list=[]
    for i in range(1,6,1):
        x = df[var].unique()[i-1] if df[var].nunique() >= i else ''
        list.append(x)
    fig = px.line(df_chart, 
        y='Reviews',   
        x='at_ym',
        title="Monthly Number of Reviews by " + var.capitalize(),
        labels={'at_ym': 'Review Month'},
        text='Reviews',
        template='ggplot2',
        hover_data={"at_ym": "|%B %d, %Y"},
        color=var,
        color_discrete_map={
            list[0]: c1, 
            list[1]: c2, 
            list[2]: c3, 
            list[3]: c4,     
            list[4]: c5
            })
    fig.update_traces(textposition='top center')
    fig.update_layout(legend_title_text=var.capitalize())
    fig.update_xaxes(
        dtick="M1",
        tickformat="%B\n%Y")
    fig.show()

```

```python
df_agg = df.groupby(['Source']).agg( 
    count_review=('Reviews', np.sum),
    score_mean=('score', 'mean'),
    first_review=('at', 'min'),
    last_review=('at', 'max'),
    versions= ('appVersion', pd.Series.nunique),
    count_reply=('reply', np.sum),
    ).reset_index().T
```
