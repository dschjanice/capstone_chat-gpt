```python
countries_lower = []
for i in range(0, len(countries)):
    c = countries[i].lower()
    countries_lower.append(c)
```

```python

dfs = pd.DataFrame()

for country_code in countries_lower:
    app_store = AppStore(country=country_code, app_name='chatgpt', app_id=6448311069)
    app_store.review(how_many=50000, sleep=25)
    df = pd.DataFrame(app_store.reviews)
    
    df['country'] = country_code
    
    dfs = pd.concat([dfs, df], axis = 0)
    time.sleep(120)
```
